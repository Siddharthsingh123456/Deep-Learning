# Assignment 4: Self-Organizing Maps (SOM) for Customer Segmentation

Complete notebook structure for Customer Segmentation using Self-Organizing Maps (Kohonen Networks).

```

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from scipy.stats import f_oneway
from minisom import MiniSom
```

## Load Dataset and EDA

```

df = pd.read_csv('credit_card_customers.csv')
df.head()
```

## Missing Values, Outlier Detection, Feature Engineering

```

print(df.isnull().sum())
```

## Scaling Comparison (MinMax vs Standardization)

```

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df.select_dtypes(include=np.number))
```

## PCA Visualization

```

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
plt.scatter(X_pca[:,0], X_pca[:,1])
plt.show()
```

## SOM Training

```

som = MiniSom(
    x=10,
    y=10,
    input_len=X_scaled.shape[1],
    sigma=3,
    learning_rate=0.5
)

som.random_weights_init(X_scaled)
som.train_random(X_scaled, 10000)
```

## Cluster Identification using K-Means

```

weights = som.get_weights().reshape(-1, X_scaled.shape[1])
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(weights)
```
